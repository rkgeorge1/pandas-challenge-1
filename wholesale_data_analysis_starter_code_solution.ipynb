## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
client_dataset_df['line_subtotal'] = client_dataset_df['unit_price'] * client_dataset_df['qty']
client_dataset_df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
client_dataset_df['shipping_price'] = client_dataset_df.apply(
    lambda row: row['unit_weight'] * row['qty'] * 7 if row['unit_weight'] * row['qty'] > 50 
    else row['unit_weight'] * row['qty'] * 10,
    axis=1
)
client_dataset_df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
client_dataset_df['total_price'] = (client_dataset_df['line_subtotal'] + client_dataset_df['shipping_price']) * 1.0925
client_dataset_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).

client_dataset_df['line_cost'] = (client_dataset_df['unit_cost'] * client_dataset_df['qty']) + client_dataset_df['shipping_price']
client_dataset_df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
client_dataset_df['line_profit'] = client_dataset_df['total_price'] - client_dataset_df['line_cost']
client_dataset_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

top_5_clients_spend = client_dataset_df[client_dataset_df['client_id'].isin([33615, 66037, 46820, 38378, 24741])].groupby('client_id')['total_price'].sum()
top_5_clients_spend
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_df = client_dataset_df[client_dataset_df['client_id'].isin([33615, 66037, 46820, 38378, 24741])].groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('line_profit', 'sum')
)
summary_df
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 

# Define a function that converts a dollar amount to millions.

# Apply the currency_format_millions function to only the money columns. 

# Rename the columns to reflect the change in the money format. 

# Formatting and defining money columns to millions
formatted_summary_df = summary_df.rename(columns={
    'total_units_purchased': 'Total Units Purchased',
    'total_shipping_price': 'Total Shipping Cost',
    'total_revenue': 'Total Revenue ($)',
    'total_profit': 'Total Profit ($)'
})

# Readability formatting
formatted_summary_df['Total Units Purchased'] = formatted_summary_df['Total Units Purchased'].apply(lambda x: f"{x:,}")
formatted_summary_df['Total Shipping Cost'] = formatted_summary_df['Total Shipping Cost'].apply(lambda x: f"${x:,.2f}")
formatted_summary_df['Total Revenue ($)'] = formatted_summary_df['Total Revenue ($)'].apply(lambda x: f"${x:,.2f}")
formatted_summary_df['Total Profit ($)'] = formatted_summary_df['Total Profit ($)'].apply(lambda x: f"${x:,.2f}")

# Convert to millions
def currency_format_millions(value):
    return f"${value / 1_000_000:.2f}M"

formatted_summary_df['Total Shipping Cost'] = formatted_summary_df['Total Shipping Cost'].replace({'\$': '', ',': ''}, regex=True).astype(float)
formatted_summary_df['Total Revenue ($)'] = formatted_summary_df['Total Revenue ($)'].replace({'\$': '', ',': ''}, regex=True).astype(float)
formatted_summary_df['Total Profit ($)'] = formatted_summary_df['Total Profit ($)'].replace({'\$': '', ',': ''}, regex=True).astype(float)

formatted_summary_df['Total Shipping Cost'] = formatted_summary_df['Total Shipping Cost'].apply(currency_format_millions)
formatted_summary_df['Total Revenue ($)'] = formatted_summary_df['Total Revenue ($)'].apply(currency_format_millions)
formatted_summary_df['Total Profit ($)'] = formatted_summary_df['Total Profit ($)'].apply(currency_format_millions)

# Rename columns to reflect the new format
formatted_summary_df = formatted_summary_df.rename(columns={
    'Total Shipping Cost': 'Total Shipping Cost (Millions)',
    'Total Revenue ($)': 'Total Revenue (Millions $)',
    'Total Profit ($)': 'Total Profit (Millions $)'
})

formatted_summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
```

```python
sorted_summary_df = formatted_summary_df.sort_values(by='Total Profit (Millions $)', ascending=False)
sorted_summary_df

```
